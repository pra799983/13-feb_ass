```python
Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.
```

```python
In Python, the Exception class is the base class for all built-in and custom exceptions. When we create a custom exception,
we typically inherit from the Exception class to create a new exception type that has specific behaviors or properties.

There are several reasons why we use the Exception class as a base class when creating custom exceptions:

0. Inheritance: By inheriting from the Exception class, our custom exception automatically inherits all the properties and
   behaviors of the Exception class. This includes the ability to add custom error messages, tracebacks, and 
   other information that can help debug issues in our code.

1. Compatibility: By inheriting from the Exception class, our custom exception is compatible with all existing code that expects to handle exceptions 
   of type Exception. This allows us to use our custom exception seamlessly in any code that handles exceptions.

2. Standardization: By using the Exception class as a base class, we adhere to the standard practice for creating exceptions in Python. This makes our
   code more readable and easier to maintain for other developers who may be working on the same codebase.

3. Customization: By inheriting from the Exception class, we can customize our custom exception to suit our specific needs. We can add new properties 
   or methods that are specific to our custom exception, and override existing methods to provide custom behavior when the exception is raised.

In summary, using the Exception class as a base class when creating custom exceptions provides many benefits, including inheritance, compatibility, 
standardization, and customization. It is considered best practice to use the Exception class as a base class when creating custom exceptions in
Python.
```

```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
```

```python

In Python, the ArithmeticError class is a base class for all exceptions that occur during arithmetic operations.
It is a subclass of the Exception class and a superclass of specific arithmetic exceptions like ZeroDivisionError, OverflowError,
and FloatingPointError. The ArithmeticError class itself does not define any specific error types, but rather serves
as a general category for arithmetic-related exceptions.
```

```python
## 1. ZeroDivisionError
try:
    15/0
except ZeroDivisionError as e:
    print(e)
```

```python
## 2.OverflowError
import sys
a = sys.maxsize
b = a + 1  # OverflowError: Python int too large to convert to C long
```

```python
The LookupError class is a base class for all exceptions that occur when a key or index lookup fails. It is a subclass of the Exception class and
is itself a superclass of specific lookup exceptions like KeyError and IndexError. The LookupError class is used as a general category for 
lookup-related exceptions, and it is often used to catch and handle errors that occur during dictionary lookups or list indexing.
```

```python
## example 1
my_dict = {'apple': 1, 'banana': 2, 'orange': 3}
value = my_dict['pear']  # KeyError: 'pear'


try:
    d = {"key":"pradeep", 1 :[1,2,34,5]}
    print(d["key2"])
except KeyError as e:
    print(e)
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?
```

```python
ImportError is a built-in exception in Python that is raised when a module or package cannot be imported. It is a subclass of the Exception 
class and is often used to catch and handle import-related errors.

There are several reasons why an ImportError might be raised. For example, it could be because the module or package does not exist in the 
specified location, the file has incorrect permissions, or the module is not installed.
```

```python
import logging
logging.basicConfig(filename  = "error.log",level = logging.ERROR)
try:
    import my_module
except ModuleNotFoundError as e :
    logging.error("The module 'my_module' could not be found. {}".format(e))
```
